```python
import pandas as pd
import seaborn as sns
url = "https://raw.githubusercontent.com/KaimwenK/MKT-440-of-Wen/main/Chipotle.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
# Select the continuous variables
cont_vars = df.select_dtypes(include=["float64", "int64"])
```

```python
# Descriptive stats for the continuous variables
cont_vars.describe()
```

```python
# pairplot of the continuous variables
sns.pairplot(cont_vars)
```

```python
#correlation matrix
corr = cont_vars.corr()
```

```python
# heatmap of the correlation matrix
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
# select the categorical variables
categorical = df.select_dtypes(include=["object"])
```

```python
categorical
```

```python
# Use seaborn to create a countplot for each categorical variable
import matplotlib.pyplot as plt
fig, ax =plt.subplots(2,3)
for col in categorical:
    sns.countplot(x=col, data=df)
fig.show()
```

```python
#include only the importance-related variables for the heatmap
#select the importance-related variables
importance = df.iloc[:, 7:12]
```

```python
#correlation matrix
corr = importance.corr()
```

```python
#create a heatmap of the correlation matrix
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
#Create one scatterplot with age as X-axis and income as Y axis, gender as hue, using seaborn library
sns.scatterplot(x="age", y="income", hue="gender", data=df)
```
